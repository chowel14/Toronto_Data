```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

Convert table to Pandas DF for cleaning and preprocessing

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

Data preprocessing and cleaning

```python
# Drop rows where Borough is NA
df1 = df[df.Borough != 'Not assigned']

# Combine the neighborhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replace the name of the neighborhoods which are NA with names of boroughs
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])

df2
```

```python
df2.shape
```

```python
df2.rename(columns={'Postal Code':'Postcode'},inplace=True)
df2.head()
```

Import the csv file containing the latitudes and longitudes for various neighborhoods in CA

```python
latlon=pd.read_csv('https://cocl.us/Geospatial_data')
latlon.head()
```

Merge the two tables for getting the latitudes and longitudes for neighborhoods in CA

```python
latlon.rename(columns={'Postal Code':'Postcode'},inplace=True)
latlon.head()
```

Select all rows from the data frame which contains Toronto in their borough

```python
df3=pd.merge(df2,latlon,on='Postcode')
df3.head()
```

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

Visualize all neighborhoods of the above data frame using Folium


```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

Use Kmeans clustering for the clustering of neighborhoods

```python
k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4
```
